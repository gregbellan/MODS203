# Data construction

## Imports

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

"https://steamcommunity.com/robots.txt"
```

## Steam data scrapping

```python
# test with one video game page
url = "https://store.steampowered.com/app/582160/Assassins_Creed_Origins/"

r=requests.get(url)

if r.status_code==200:
    html=r.text
    print(html)
else:
    print("error : page not found")
```

```python
# initialize data frame
df=pd.DataFrame()

# initialize the id_game of the selected games
list = [397540, 582160, 1174180]


for i in list :
    # extract url of game page
    token = "https://store.steampowered.com/app/"
    url = token + str(i) 
    r=requests.get(url)
    if r.status_code==200:
        html=r.text
    else:
        print("error : page not found")
        
    # initialize game line
    game = {"id_game" : i, "name" : None, "nb Overall Reviews" : None, "nb Recent Reviews" : None}
    
    # extract data from page
    bs = BeautifulSoup(html, 'html.parser')
    game["name"] = bs.find('div', class_= "details_block").findAll('b')[0].next_sibling
    reviews = bs.findAll('div', class_= "user_reviews_summary_bar")
    game["nb Overall Reviews"] = int(re.sub('\D', '', reviews[0].get_text()))
    game["nb Recent Reviews"] = int(re.sub('\D', '', reviews[1].get_text()))
    df = df.append(game, ignore_index=True)
    
print(df)
df.to_csv(index = False, header=True)
```
